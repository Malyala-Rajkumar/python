```python
#1)Write a Python function that takes a list of numbers as input and returns a sorted version of the list without using any built-in sorting functions.

def withOutBuiltInFun(arr):
    """sorting the list with out using the inbuild functions"""
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


# calling the function
withOutBuiltInFun([4, 5, 9, 10, 1])
```

```python
"""5) text = "Data science is amazing"
-> Count the number of vowels (a, e, i, o, u) in each word.
-> Return a dictionary with word as key and vowel count as value."""


def count_Vowels(text):
    """count the vowels each of the word in the given sentence"""
    vowels = 'aeiou'
    results = {}
    for word in text.split():
        count = sum(1 for ch in word.lower() if ch in vowels)
        results[word] = count
    return results


count_Vowels("Data science , is amazing")
```

```python
"""6) Find the sum of the indices of the second largest and second smallest numbers in the list 
x = ['24', '3', '14', '36', '5']"""


def sum_indices(arr):
    sorted_arr = sorted(arr)
    second_small = sorted_arr[1]
    second_large = sorted_arr[-2]
    return arr.index(second_small) + arr.index(second_large)

print('Sum of the indices of 2 values is:- ', sum_indices(['24', '3', '14', '36', '5']))
```

```python
"""8) Check if a number is an Armstrong number using a while loop.
Example: 153 → 1³ + 5³ + 3³ = 153"""

num = int(input())
num_copy = num
out = 0

while num > 0:
    rem = num % 10
    out = out + (rem ** 3)
    num = num // 10
    
if num_copy == out:
    print(f"Yes the Number '{num_copy}' is Armstrong Number")
else:
    print(f"No the Number '{num_copy}' is Armstrong Number")
```

`What is the difference between @staticmethod and @classmethod`:- 

    @staticmethod

    - whatever the function in the class that calss can be converted to the the static metthod by using the @staticmethod
    - A static method does not receive an implicit first argument.
    
        class Method:
            @staticmethod
            def arg(**kwargs):
                pass
    
    - it can be called either the class or we can call the instance of the class
    
    @classmethod
    
    - A class method receives the class as implicit first argument, just like an instance method receives the instance.
    -  It can be called either on the class or an instance. 
    - The instance is ignored except for its class.If a class method is called for a derived class, the derived class object is passed as the implied first argument.
    
    class Method:
            @classmethod
            def arg(**kwargs):
                pass

```python
# 4
def Capitalize_ASSCII(s):
    empty = " "
    for ch in s:
        if ord(ch) % 2 == 0:
            empty += ch.lower()
        else:
            empty += ch.upper()
    return empty


Capitalize_ASSCII("Rajkuamr Malyala")
```

```python
# 5️⃣ Index difference between duplicate elements

arr = [1,2,3,1,2,4]

result = []

for i in range(len(arr)):
    found = False

    for j in range(i+1,len(arr)):
        if arr[i] == arr[j]:
            result.append(j-i)
            found = True
            break

    if not found:
        result.append(0)

print(result)
```

```python
# 6️⃣ Shopping Cart using OOP

class ShoppingCart:

    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(item,"added")

    def remove_item(self,item):
        if item in self.items:
            self.items.remove(item)

    def display(self):
        print("Cart Items:", self.items)
        print("Total items:", len(self.items))


cart = ShoppingCart()

cart.add_item("Laptop")
cart.add_item("Phone")
cart.add_item("Headphones")

cart.display()

cart.remove_item("Phone")

cart.display()
```

```python
# 9️⃣ First non-repeating character

def first_non_repeating(s):

    count = {}

    for ch in s:
        count[ch] = count.get(ch,0) + 1

    for ch in s:
        if count[ch] == 1:
            return count.items() , ch

print(first_non_repeating("aabbcdde"))
```

```python
# Method 2 (Using collections)
from collections import Counter

def first_unique(s):

    c = Counter(s)

    for i in s:
        if c[i] == 1:
            return i

print(first_unique("aabbcdde"))
```
