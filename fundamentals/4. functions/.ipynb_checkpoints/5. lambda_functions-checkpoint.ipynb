### Python Lambda Functions

    Lambda functions are small anonymous functions, meaning they do not have a defined name. In Python, lambda functions are created using the lambda keyword for short, simple operations. 

    - No name unless you assign it to a variable
    - Only one expression
    - Implicit return of that expression

`Lambdas are small, short-lived functions created and used immediately, mainly to pass simple logic as an argument to another function.`

```python
s = 'Rajkumar Malyala'
upper = lambda x:x.upper()
upper(s)
```

```python
students = [('Raj',25), ('John',20), ('Sam',30)]

sorted_students = sorted(students, key=lambda x: x[1])
sorted_students
```

Syntax of Lambda Expression

<center>  </center>

- Function name (a): Stores the lambda function so it can be reused later.
- Lambda keyword (lambda): Defines an anonymous (inline) function in Python.
- Argument (x): The input value passed to the lambda function.
- Expression (x**2): The operation performed on the argument and returned as the result.

#### Use Cases of Lambda Functions

```python
"""
1. Using with Condition Checking
A lambda function can use conditional expressions (if-else) to return different results based on a condition.
"""

check = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(check(5))
print(check(-3))
print(check(0))
```

```python
"""2. Using with List Comprehension
Lambda functions can be combined with list comprehensions to apply the same operation to multiple values in a compact way."""

func = [lambda arg=x: arg * 10 for x in range(1, 5)]
for i in func:
    print(i())
```

```python
"""
3. Using for Returning Multiple Results
Although a lambda can contain only one expression, it can still return multiple results by combining them into a tuple.
"""

calc = lambda x, y: (x + y, x * y)
res = calc(3, 4)
print(res)
```

```python
"""
4. Using with filter()
filter() function uses a lambda expression to select elements from a list that satisfy a given condition, such as keeping only even numbers.
"""

c = [1, 2, 3, 4, 5, 6]
even = filter(lambda x: x % 2 == 0, c)
print(list(even))
```

```python
"""
5. Using with map()
map() function applies a lambda expression to each element of a list and returns a new list with the transformed values
"""

a = [1, 2, 3, 4]
double = map(lambda x: x * 2, a)
print(list(double))
```

```python
"""
6. Using with reduce()
reduce() function repeatedly applies a lambda expression to elements of a list to combine them into a single result.
"""

from functools import reduce
a = [1, 2, 3, 4]
mul = reduce(lambda x, y: x * y, a)
print(mul)
```

#### Functions

    filter() in python

    filter() function is used to extract elements from an iterable (like a list, tuple or set) that satisfy a given condition. It works by applying a function to each element and keeping only those for which function returns True.

    Iterable:- An iterable is an object that can return its elements one by one using an iterator.
    Iterator:- An iterator is an object that produces elements one by one using the next() function.

```python
for x in numbers:
    print(x)
    
iterator = iter(numbers)

while True:
    try:
        x = next(iterator)
        print(x)
    except StopIteration:
        break
```

```python
# Basic Example how to keep only the words starting with the letter 'a' from a list of fruits.

def starts_a(w):
    return w.startswith("a")

li = ["apple", "banana", "avocado", "cherry", "apricot"]
res = filter(starts_a, li)
print(list(res))
```

```python
# Examples of filter function
```

```python
"""Example 1: Using filter() with a Named Function
This code defines a regular function to check if a number is even and then uses filter() to extract all even numbers from a list."""

def even(n):
    return n % 2 == 0


a = range(1,11)
b = filter(even, a)
list(b)
```

```python
"""Example 2: Using filter() with a Lambda Function
Instead of creating a separate named function, use a lambda function for concise code. Below code uses a lambda function with filter() to select even 
numbers from a list."""

a = range(1,11)
b = filter(lambda x: x % 2 == 0, a)
list(b)
```

```python
"""Example 3: Filtering and Transforming Data
In this Example lambda functions is used with filter() and map() to first get even numbers from a list and then double them."""

a = [1, 2, 3, 4, 5, 6]
b = filter(lambda x: x % 2 == 0, a)
c = map(lambda x: x * 2, b)
print(list(c))
```

```python
"""Example 4: Filtering Strings
Here, lambda function is used with filter() to keep only words that have more than 5 letters from a list of fruits.
"""

a = ["apple", "banana", "cherry", "kiwi", "grape"]
b = filter(lambda x: len(x) > 5, a)
list(b)
```

```python
"""Example 5: Filtering with None (Truthiness Check)
This code uses filter() with None as the function to remove all falsy values (like empty strings, None and 0) from a list.
"""

L = ["apple", "", None, "banana", 0, "cherry"]
A = filter(None, L)
print(list(A))
```

    Python map() function

    map() function in Python applies a given function to each element of an iterable (list, tuple, set, etc.) and returns a map object (iterator). It is a higher-order function used for uniform element-wise transformations, enabling concise and efficient code.

```python
s = ['1', '2', '3', '4']
res = map(int, s)
print(list(res))
```

```python
"""Converting map object to a list
By default, map() function returns a map object, which is an iterator. In many cases, we will need to convert this iterator to a list to work with the 
results directly.
"""


# Custom function to be applied
# in map
def double(val):
    return val * 2


# Let us apply double on every member
a = [1, 2, 3, 4]
res = list(map(double, a))
print(res)
```

```python
"""map() with lambda
We can use a lambda function instead of a custom function with map() to make code shorter and easier. Let's see how to improve above code for better 
readability."""

a = [1, 2, 3, 4]
res = list(map(lambda x: x ** 2, a))
print(res)
```

```python
"""map() with multiple iterables
We can use map() with multiple iterables if the function we are applying takes more than one argument."""

a = [1, 2, 3]
b = [4, 5, 6]
res = map(lambda x, y: x + y, a, b)
print(list(res))
```

```python
"""
Converting strings to Uppercase
we can use map() to convert a list of strings to uppercase.
"""

fruits = ['apple', 'banana', 'cherry']
res = map(str.upper, fruits)
list(res)
```

```python
"""
Extracting first character from strings
we use map() to extract the first character from each string in a list.
"""

words = ['apple', 'banana', 'cherry']
res = map(lambda s : s[0], words)

list(res)
```

```python
"""
Removing whitespaces from strings
We can use map() to remove leading and trailing whitespaces from each string in a list.
"""

s = ['  hello  ', '  world ', ' python  ']
res = map(str.strip, s)
print(list(res))
```

```python
"""
Calculate fahrenheit from celsius
we use map() to convert a list of temperatures from Celsius to Fahrenheit.
"""

celsius = [0, 20, 37, 100]
fahrenheit = map(lambda c: (c * 9/5) + 32, celsius)
print(list(fahrenheit))
```

    reduce() in Python

    The reduce() function in Python (from the functools module) applies a function cumulatively to the elements of an iterable and returns a single final value. It processes elements step-by-step, combining two elements at a time until only one result remains.

```python
from functools import reduce
a = ["Geeks", "for", "Geeks"]
r = reduce(lambda x, y: x + " " + y, a)
print(r)
```

```python
# Example 1: Here, the code adds all numbers in a list using reduce(). The function combines two numbers at a time.

from functools import reduce
a = [2, 4, 6, 8]
r = reduce(lambda x, y: x + y, a)
print(r)
```

```python
# Example 2: Here, the code finds the largest number from a list using reduce().

from functools import reduce
a = [5, 9, 3, 12, 7]
r = reduce(lambda x, y: x if x > y else y, a)
print(r)
```

```python
# This example uses functools.reduce() with built-in functions from operator module to perform sum, product and string concatenation on lists.

import functools
import operator
a = [1, 3, 5, 6, 2]


print(functools.reduce(operator.add, a))
print(functools.reduce(operator.mul, a)) 
print(functools.reduce(operator.add, ["geeks", "for", "geeks"]))
```

```python
""" This code demonstrates how accumulate() from itertools module works it performs cumulative operations and returns all intermediate results instead 
of just a single final value."""

from itertools import accumulate
from operator import add


a = [1, 2, 3, 4, 5]
res = accumulate(a, add)
print(list(res))
```

    lambda vs def Keyword

    In Python, both lambda and def can be used to define functions, but they serve slightly different purposes. While def is used for creating standard reusable functions, lambda is mainly used for short, anonymous functions that are needed only temporarily.

```python
# Using lambda
square = lambda x: x ** 2
print(square(3))

# Using def
def sqdef(x):
    return x ** 2
print(sqdef(3))
```

<center></center>

map() and filter() return iterators to support lazy evaluation. Instead of computing and storing all results in memory at once, they generate elements one by one when needed. This improves memory efficiency and allows processing of large datasets.
